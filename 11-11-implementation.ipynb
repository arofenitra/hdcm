## Components of the dataset
### 1. The prompt or instruction
It is the original language problem description (from MBPP for example) 
### 2. The generated code (with hallucination or not)
It is a python code snippet that attempt to to solve the problem but contains a deliberately introduced "hallucination" or error.
### 3. The Label/Annotation
It is A binary label (0 for correct/non-hallucination, 1 for hallucination/incorrect) and a fine-grained label for the type of hallucination (e.g., Logical Error, Invalid API Call, etc.).
